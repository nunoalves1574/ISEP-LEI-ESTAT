```python
# EXERCICIO 52
from scipy import stats
import math

# 52.1
# X: Durabilidade de um bidão, em dias
# P(X>50) = 0.1 <=> e^50.lambda = 0.1 <=> ln(0.1) = -50.lambda <=> ln(0.1) / -50 = lambda 
lambda1 = math.log(0.1) / (-50)
beta = 1/lambda1
print(f'A durabilidade esperada por bidão é: {beta:.1f}')

# 52.2
# P(X>30)
result = 1 - stats.expon.cdf(30, scale = beta)
print(f'A percentagem de bidões que mantêm o quimico por mais de 30 dias é {result*100:.1f} %')

# 52.3
# P(X<10) 
p10 = stats.expon.cdf(10, scale = beta)
#P(10<X<30)
p10_30 = p10 - p30
# P(X>30) 
p30 = result
print(f'p10 = {p10} ; p1030 = {p10_30} ; p30 = {p30}')
valor_esperado = 5*p10 + 30*p10_30 + 100*p30
print(f'O valor esperado a receber por bidão é: {valor_esperado:.1f}')
```

```python
# EXERCICIO 62
# T: Peso de uma caixa cheia, em gramas
# Xi: Peso do i-ésimo pires, em gramas
# Xi ~ N(190g, 100g)
# Yi: Peso da i-ésima chávena, em gramas
# Yi ~ N(170g, 225g)

# 62.1
# Queremos P(T < 2000)
# T = somatório 5 de Xi + somatório 5 de Yi + 100 ~ N(media, variancia)
# Média T = 5 * media Xi + 5 * media Yi + 100 = 1900
# Variancia T = 5 * variancia Xi + 5 * variancia Yi + 0(variancia de 100, é uma constante) = 1625
# Desvio padrao T = raiz qudrada de 1625 = 40.3
result = stats.norm.cdf(2000, 1900, 40.3)
print(result)

# 62.2
# X: Peso de um pires, em gramas
# X ~ N(190g, 100g)
# Y: Peso de uma chávena, em gramas
# Y ~ N(170g, 225g)
# P(Y>X) = P(Y - X > 0) = P(W > 0) = 1 - P(W <= 0)
# W = Y - X
# média W = média de Y - média de X
# varianca W = variancia de Y + variancia de X
# W ~ N(-20, 325) 

result = 1 - stats.norm.cdf(0, -20, 18.03)
print(result)
```

```python
# EXERCICIO 66
from scipy import stats
import numpy as np
# X: Peso de açucar numa chávena, em g
# X ~ N(70, 3)

# Y: Peso de uma clara, em g
# Y ~ N(14, desvio padrao)

# P(Y<15) = 0.7
# P(Z<z) = 0.7
Z = stats.norm.ppf(0.7, 0, 1)
desvioPadrao = (15-14)/Z
# Y ~N(14, (15-14)/Z)

# X2: Peso de açucar em 2 chávenas, em g
# X2 ~ N(140, 18)
# Y8: Peso de oito claras, em g
# Y8 ~ N(112, desviopadrao^2 * 8)

varianciaClaras = desvioPadrao * desvioPadrao * 8
variancia = 18 + varianciaClaras
desvioPadrao = np.sqrt(variancia)
print(f'{desvioPadrao:.3f}')
# T: Peso da massa de um doce, em g
# T ~ N(252, 18+varianciaClaras)

# P(T>240)
result = 1 - stats.norm.cdf(240, 252, desvioPadrao)
print(f'A probabilidade de o peso da massa do doce ser superior a 240g é: {result:.3f}')

# 66.2
# P(Y>15 | Y > 14) = P(Y>15) / Y > 14) = 0.3/0.5
clarasAcima15 = 0.3/0.5
print(f'{clarasAcima15:.3f}')



```
