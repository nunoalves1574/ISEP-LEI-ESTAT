```python
# EXERCICIO 5
from scipy import stats
import numpy as np

# 5.1
# A ~ N(10, 4)
mediaA = 10
desvioA = 2

# B ~ Uni(5,19)
# P(B<6) probabilidade do aspirador avariar antes dos 6 meses
probAvariarB = stats.uniform.cdf(6, 5, 14)

lucroUniB = 2100 # lucro unitario
prejuizoUniB = -7000 # prejuizo unitario

# probabilidade de avariar apos os 6 meses * o lucro
lucroB = (1 - probAvariarB) * lucroUniB

# probabilidade de avariar antes os 6 meses * o prejuizo
prejuizoB = probAvariarB * prejuizoUniB

# calcular media de lucro/prejuizo
mediaLucroB = lucroB + prejuizoB

# variancia = E[X**2] - E[X]**2
varianciaLucroB = (((1 - probAvariarB) * lucroUniB**2) + (probAvariarB * prejuizoUniB**2) - mediaLucroB**2)
print(f'{varianciaLucroB:.2f}')


# 5.2
# para calcular o lucro medio
# P(A<6)
probAvariarA = stats.norm.cdf(6, 10, 2)

lucroUniA = 1200 # lucro unitario
prejuizoUniA = -2500 # prejuizo unitario

# probabilidade de avariar apos os 6 meses * o lucro
lucroA = (1 - probAvariarA) * lucroUniA

# probabilidade de avariar antes os 6 meses * o prejuizo
prejuizoA = probAvariarA * prejuizoUniA

# calcular media de lucro/prejuizo
mediaLucroA = lucroA + prejuizoA

# variancia = E[X**2] - E[X]**2
varianciaLucroA = (((1 - probAvariarA) * lucroUniA**2) + (probAvariarA * prejuizoUniA**2) - mediaLucroA**2)

# A - B ~ N(mediaLucroA - mediaLucroB, varianciaLucroA/50 + varianciaLucroB/40)

mediaAmostral = mediaLucroA - mediaLucroB
varianciaAmostral = (varianciaLucroA / 50) + (varianciaLucroB / 40)
result = 1 - stats.norm.cdf(0, mediaAmostral, np.sqrt(varianciaAmostral))
print(f'Media Amostral: {mediaAmostral}')
print(f'Desvio Amostral: {np.sqrt(varianciaAmostral)}')
print(f'A probabilidade de o lucro de A ser maior do que o lucro de B: {result:.2f}')
```

```python
# EXERCICIO 12
from scipy import stats
import numpy as np

# Xa : "Nº de artigos vendidos no posto A numa semana"
# Xa ~ binom(5, 0.4)
mediaA = 5 * 0.4
varianciaA = 5 * 0.4 * 0.6
desvioA = np.sqrt(varianciaA)

# Xb : "Nº de artigos vendidos no posto B numa semana"
nB = 100
mediaAmostralB = 1/nB * (0*10 + 1*25 + 2*35 + 3*30)

# Calcula-se a variancia amostral com base a formula de S^2 visto o desviopadrao e a variancia de B ser desconhecido
# S^2 = 1/(n-1) * [(sumatorio xi^2) - (n*(mediaamostral**2))]
varianciaAmostralB = 1/(nB-1) * ((0**2 * 10 + 1**2 * 25 + 2**2 * 35 + 3**2 * 30) - nB * (mediaAmostralB**2))
desvioAmostralB = np.sqrt(varianciaAmostralB)

# 1 - alfa = 0.9 <=> alfa = 0.1, contudo na pratica é utilizado Z 1 - alfa/2 <=> 1 - 0.05 = 0.95
# Logo P(Z < z) = 0.95 e Z ~ N(0,1)
valorCritico = stats.norm.ppf(0.95, 0, 1)
ICmenor = mediaAmostralB - valorCritico*(desvioAmostralB / np.sqrt(100))
ICmaior = mediaAmostralB + valorCritico*(desvioAmostralB / np.sqrt(100))
print(f'O intervalo de confiança a 90% para a média do posto B é de {ICmenor:.3f} a {ICmaior:.3f}')
print('Como o valor de mediaA = 2, está dentro do intervalo de confiança, logo não ha evidências de diferença.')

# 12.2
# regiao critica Z < -valorCritico ou Z > valorCritico
# ou seja, Z < -1,645 ou Z > 1,645
# sendo Z = mediaAmostralB - mediaA / (desvioAmostralB/raiz de n)
# mas como queremos o tamanho minimo de amostra temos de encontrar |Z| = valorCritico
# |mediaAmostralB - mediaA / (desvioAmostralB/raiz de n)| = valorCrtitico
# |mediaAmostralB - mediaA| * raiz de n / desvioAmostralB = valorCritico
# |mediaAmostralB - mediaA| * raiz de n = valorCritico * desvioAmostralB
# raiz de n = valorCritico * desvioAmostralB / |mediaAmostralB - mediaA|
# n = (valorCritico * desvioAmostralB / |mediaAmostralB - mediaA|)**2

n = ((valorCritico * desvioAmostralB) / (mediaAmostralB - mediaA))**2
print(f'O tamanho minimo de amostra deveria ser considerado para que a resposta da na alinea 12.1 fosse contraria era: {n:.0f}')

# 12.3
# para descobrirmos o nivel de confiança máxima de forma a contradizer o 12.1
# calculamos o valor de |Z| com os dados que fomos obtendo ...
# ... Z = mediaAmostralB - mediaA / (desvioAmostralB/raiz de n) ...
# ... (em vez de usarmos os 90% de confiança aplicados em 1-alfa/2 na 12.1)
Z = abs((mediaAmostralB - mediaA) / (desvioAmostralB / np.sqrt(nB)))
# e agora o valor de P(Z < z)
p = stats.norm.cdf(Z, 0, 1)
# e agora o valor de alfa invertendo a formula ou seja 1 - alfa/2 = p <=> 1 - p = alfa/2 <=> 2 * (1-p) = alfa
alfa = 2 * (1 - p)
confianca = 1 - alfa
print(f'{confianca*100:.2f}%')
```

```python
# EXERCICIO 19
from scipy import stats
import numpy as np

# 19.1
# proporção ~ N (p; p(1-p) / n) ~ N (0.45; 0.45 * 0.55 / 200)
n = 200
variancia = (0.45 * 0.55) / n

# a maioria da população ser contra é P(P^>0.5)

result = 1 - stats.norm.cdf(0.5, 0.45, np.sqrt(variancia))
print(f'{result:.3f}')

# 19.2
# p^ = 125/200 = 0.625
# p = P^ +/- erro
# erro = Z(1 - alfa/2) * desvioPadrao = Z(1 - alfa) * raiz de (p(1-p)/n), mas como nao sabemos p, = Z(1 - alfa) * raiz de (P^(1-P^)/n)
# como n>=30 => proporção ~ N (p; p(1-p) / n) ^ Z = P^ - p / raiz de (P^(1-P^)/n) ~ N(0,1)
# 1 - alfa = 0.98 => alfa = 0.02 => alfa/2 = 0.01
# logo usamos P(Z < z) = 0.99
p = 125/200
valorCritico = stats.norm.ppf(0.99, 0, 1)
erro = valorCritico * np.sqrt((p * (1 - p)) / n)
ICinf = p - erro
ICsup = p + erro

print(f'O intervalo de confiança para 98% é [{ICinf:.2f}, {ICsup:.2f}]')

```
