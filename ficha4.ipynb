```python
# EXERCICIO 1
from scipy import stats
import numpy as np
result = 1 - stats.norm.cdf(2490, 2500, np.sqrt(1025))
print(f'{result:.3f}')

# 1.2
result = 1 - stats.norm.cdf(205, 200, np.sqrt(33))
print(f'{result:.3f}')
```

```python
# EXERCICIO 2
from scipy import stats
# X: Nota de exame
# X ~ N(10.8, 2.02)

# 2.1
# x: Nota minima dos 1º 20%
# P(X>=x) = 0.20 <=> 1 - P(X<x) = 0.20 <=> F(X) = 0.8
result = stats.norm.ppf(0.8, 10.8, 2.02)
print(f'{result:.3f}')

# 2.2
result = stats.norm.cdf(1, 0, 0.49) - stats.norm.cdf(-1, 0, 0.49)
print(f'{result:.3f}')
```
