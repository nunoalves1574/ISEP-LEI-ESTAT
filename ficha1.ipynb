```python
#EXERCICIO 1
valorEmprestimo = float(input('Introduza o valor do empréstimo: '))
salario = float(input('Introduza o salário do comprador: '))
anos = int(input('Introduza a quantidade de anos a creditar: '))

qtdMeses = anos * 12
valorPrestacao = valorEmprestimo / qtdMeses

if (valorPrestacao < (salario * 0.3)):
    print(f'O crédito foi aprovado. O valor da prestação é {valorPrestacao} durante {anos} anos.')
else:
    print('O crédito foi recusado.')
```

```python
#EXERCICIO 2
precoCompra = float(input('Introduza o valor da compra: '))
metodoPagamento = int(input('Introduza o método de pagamento: 1 - Pronto pagamento ou 2 - Parcelado'))

if metodoPagamento == 2:
    qtdPagamentos = int(input('Introduza quantos pagamentos pretende fazer'))
    if qtdPagamentos <= 1:
        print('Introduziu um valor inválido')
    else: 
        print(f'Valor inicial: {precoCompra}')
    
    if qtdPagamentos == 2:
        print ('O preco não sofreu alterações')
    elif qtdPagamentos == 3:
        print ('O preco aumentou 20% com os juros')
        precoCompra = precoCompra * 1.2
elif metodoPagamento == 1:
    print ('Teve desconto de 10%')
    precoCompra = precoCompra * 0.9
else:
    print ('Introduziu um método errado')
    exit

print(f'O valor final é {precoCompra}')

```

```python
#EXERCICIO 3
import random
possibilidades = ['Pedra', 'Papel', 'Tesoura']
vitorias = 0
jogos = 10
for i in range (jogos):
    print('Jogo nº', i+1)
    jogada = input('Qual é a sua jogada?')

    while jogada not in possibilidades:
        jogada = int(input('Entrada inválida! Qual é a sua jogada?'))

    computador = random.choice(possibilidades)
    print('O computador jogou ', computador)

    if jogada == computador:
        print('Empate!')
    elif ( (jogada == 'Pedra' and computador == 'Tesoura') or
           (jogada == 'Papel' and computador == 'Pedra') or
           (jogada == 'Tesoura' and computador == 'Papel') ):
        print('Ganhaste!')
        vitorias+=1
    else:
        print('Perdeste!')

percentagem = vitorias / jogos * 100
print('Ganhaste ', percentagem, '% dos jogos')
```

```python
# EXERCICIO 8

lista = [['Número', 'Nome', 'Média']]

while True:
    numero = int(input('Introduza o seu número de aluno: '))
    nome = input('Introduza o seu nome: ')
    nota1 = int(input('Introduza a primeira nota: '))
    nota2 = int(input('Introduza a segunda nota: '))
    media = (nota1 + nota2)/2

    lista.append([numero, nome, media])

    continuar = input("Quer continuar? [S/N]: ").upper()
    if continuar == 'N':
        break

print("\n{:<10} {:<20} {:<10}".format("Número", "Nome", "Média"))
print("-" * 42)

for aluno in lista[1:]:
    print("{:<10} {:<20} {:<10.2f}".format(aluno[0], aluno[1], aluno[2]))
```

```python
# EXERCICIO 9

def contador(inicio, fim, passo):
    if inicio < fim:
        while inicio <= fim:
            print(inicio, end=" ")
            inicio += passo
    else:
        while inicio >= fim:
            print(inicio, end=" ")
            inicio -= passo

    print("\n")


# 1ª contagem
contador(1, 10, 1)

# 2ª contagem
contador(10, 0, 2)

start = int(input('Introduza o inicio: '))
end = int(input('Introduza o final: '))
passe = int(input('Introduza o passo personalizado: '))

contador(start, end, passe)

    

```
